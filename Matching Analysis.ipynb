```python
#Cleaning subjects data
#Input is pandas series, output is enum with strings


    
#tutors_subjects=tutors_df["What subjects are you comfortable helping your student with?"]

#copied from stack exchange
#removes things in parenthases from a string
def a(test_str):
    ret = ''
    skip1c = 0
    skip2c = 0
    for i in test_str:
        if i == '[':
            skip1c += 1
        elif i == '(':
            skip2c += 1
        elif i == ']' and skip1c > 0:
            skip1c -= 1
        elif i == ')'and skip2c > 0:
            skip2c -= 1
        elif skip1c == 0 and skip2c == 0:
            ret += i
    return ret

def clean_subjects(df):
    #get only the subjects, turn all cells into strings
    df_subjects=df.applymap(str)["Subjects"]
    #remove parenthases
    no_par_df=df_subjects.apply(a)
    #split on commas
    for index in range(1, df.shape[0]):
        df_split=no_par_df.loc[index].split(',')
        #replace with list
        df.at[index, "Subjects"]=df_split
        
clean_subjects(tutors_df)
clean_subjects(students_df)
students_df["Subjects"]
```
